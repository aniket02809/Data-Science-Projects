```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot(df['model'], df['hp'])

# Customize the plot
plt.xlabel('Models of the cars')
plt.xticks(rotation = 'vertical')
plt.ylabel('Horse-Power of Cars')
plt.title('Model Names vs Horse Power')

# Display the plot
plt.show()
```

```python
plt.bar(df["carb"],df["gear"])
plt.xlabel("Number of carburetors")
plt.ylabel("Number of forward gears")
plt.show()
```

```python
plt.hist(df['wt'], bins=30)
plt.xlabel('Weight of the cars')
plt.ylabel('Count')
plt.title('Histogram for the weight values')
plt.show()
```

```python
cyl_counts = df['cyl'].value_counts()
model_counts = df['model'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
plt.pie(cyl_counts, labels=cyl_counts.index, autopct='%1.1f%%')

# Set the title
plt.title('Distribution of Cylinders')

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.pie(model_counts, labels=model_counts.index, autopct='%1.1f%%')

# Set the title
plt.title('Distribution of Models')

# Display the plot
plt.show()
```

```python
plt.figure(figsize = (8,6))
plt.fill_between(df["am"],df["carb"], color = "skyblue")
plt.xlabel("Transmmission")
plt.ylabel("Numbers of carbureters")
plt.title("Transmission vs number of carbureters")
plt.show()
```
