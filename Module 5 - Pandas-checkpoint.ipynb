```python
import pandas as pd

def create_dataframe(keys, values):
    data = {key: value for key, value in zip(keys, values)}
    return pd.DataFrame(data)

def concatenate_dataframes(df1, df2):
    concatenated = pd.concat([df1, df2], ignore_index=True)
    return concatenated

keys1 = ["Name", "Age"]
values1 = [["John", "Alice"], [25, 30]]
df1 = create_dataframe(keys1, values1)

keys2 = ["Name", "Age"]
values2 = [["Bob", "Eve"], [35, 40]]
df2 = create_dataframe(keys2, values2)

concatenated_df = concatenate_dataframes(df1, df2)
print(concatenated_df)
```

```python
def find_highest_correlation(hp):
    df = pd.read_csv(r'C:\Users\dell\Downloads\python-Datasets (2)\python Datasets\python Datasets\cars (1) (1).csv')
    corr_matrix = df.corr()
    highest_corr_column = corr_matrix[hp].idxmax()
    return highest_corr_column
column_name = "mpg"
highest_corr_column = find_highest_correlation(column_name)
print(f"The column with the highest correlation to '{column_name}' is '{highest_corr_column}'.")
```
